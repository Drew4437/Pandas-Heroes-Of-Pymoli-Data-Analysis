```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
unique_items = purchase_data['Item ID'].nunique()
average_price = purchase_data['Price'].mean()
total_purchase = len(purchase_data)
total_revenue = purchase_data['Price'].sum()
total_revenue
summerize_df = pd.DataFrame({
   "Unique Items": [unique_items],
   "Average Price": [average_price],
   "Total Purchases": [total_purchase],
   "Total Revenue": [total_revenue]
})
pd.options.display.float_format = '${:,.2f}'.format
summerize_df
```

# Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#check column names
#purchase_data.columns

df1 = purchase_data.groupby(['Gender']).agg({'SN' : 'nunique' , 'Price' : 'sum'})
s = purchase_data.drop_duplicates(['Gender']).groupby(['Gender'])['Purchase ID'].sum()
df = df1.join(s.rename('Purchase ID'))
df = pd.DataFrame(df)
df

total_gender = df['SN']


total_gender = pd.DataFrame(total_gender)
total_gender.columns = ["Total Count"]
# total_gender

total = purchase_data['SN'].nunique()
# total
gender_percent = ((total_gender/total))
# gender_percent
total_gender['Percentage of Players'] = gender_percent
# total_gender

#final formatting
total_gender['Percentage of Players'] = total_gender['Percentage of Players'].astype(float).map("{:.2%}".format)
total_gender = total_gender.sort_values(by = ["Total Count"], ascending=False)
total_gender
```

# Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#check column names
purchase_data.columns

top_analysis = pd.DataFrame(purchase_data["SN"].value_counts())
top_analysis
top_analysis4 = pd.DataFrame(purchase_data)
top_analysis4 = pd.DataFrame(top_analysis4)
top_analysis4 = top_analysis4 .groupby("SN").sum()
top_analysis4


top_analysis.columns = ["Purchase Count"]
top_analysis.head()


# #!!!!average purprice
top_analysis['Average Purchase Price'] = top_analysis4['Price'] / top_analysis['Purchase Count']
top_analysis.head()


top_analysis['Total Purchase Value'] =  top_analysis['Average Purchase Price'] * top_analysis["Purchase Count"]
top_analysis


# pd.options.display.float_format = '${:,.2f}'.format
top_analysis = top_analysis.sort_values(by = ["Total Purchase Value"], ascending=False)
top_analysis.head()
```
